```python
# Synthetic Industrial Data Generation
def generate_factory_data():
    """Generates reliable, non-PII industrial data for demo purposes."""
    operations = ['Front Pocket', 'Back Yoke', 'Inseam', 'Side Seam', 'Waistband']
    
    # Line Data: Real-time status of 5 operations
    line_data = {
        'op_id': [f"OP_{i:02d}" for i in range(1, 6)],
        'operation': operations,
        'target_sam': [1.2, 0.8, 1.5, 2.0, 1.1], # Standard Allowed Minutes
        'live_sam': [1.3, 0.85, 2.2, 1.9, 1.15], # Current observed SAM
        'operator_id': ['EMP_101', 'EMP_102', 'EMP_103', 'EMP_104', 'EMP_105'],
        'safety_level': [4, 3, 4, 2, 5] # Required certification level for the machine
    }
    
    # Skill Matrix: Operator proficiency (0-100)
    skills = {
        'emp_id': ['EMP_101', 'EMP_102', 'EMP_103', 'EMP_104', 'EMP_105', 'EMP_FLOAT_01'],
        'Front Pocket': [95, 40, 60, 30, 80, 88],
        'Inseam': [50, 30, 85, 40, 70, 92],
        'cert_level': [4, 3, 4, 3, 5, 5] # Skill/Safety certification
    }
    
    return pd.DataFrame(line_data), pd.DataFrame(skills)

line_df, skill_df = generate_factory_data()
display(line_df.head())
```

```python
# The Denim-Flow Agent Logic
class DenimFlowOrchestrator:
    def __init__(self, line_data, skill_data):
        # Using .copy() to preserve the original 'Truth' data (Memory Management)
        self.state = line_data.copy()
        self.skills = skill_data.copy()
        self.logs = []

    def security_guardrail(self, emp_id, required_level):
        """Cybersecurity/Safety Node: Prevents unauthorized machine assignment."""
        emp_cert = self.skills.loc[self.skills['emp_id'] == emp_id, 'cert_level'].values[0]
        return emp_cert >= required_level

    def sam_monitor_agent(self):
        """Node 1: Efficiency Analysis (Vectorized for low latency)"""
        self.state['efficiency'] = (self.state['target_sam'] / self.state['live_sam']) * 100
        bottlenecks = self.state[self.state['efficiency'] < 85].copy()
        return bottlenecks

    def workforce_optimizer_agent(self, bottlenecks):
        """Node 2: Logic-based Re-assignment"""
        decisions = []
        for _, row in bottlenecks.iterrows():
            op_name = row['operation']
            req_safety = row['safety_level']
            
            # Find best available operator not currently in a bottleneck
            potential = self.skills.sort_values(by=op_name, ascending=False)
            
            for _, tech in potential.iterrows():
                # Security Check: Skill + Safety Certification
                if tech[op_name] > 85 and self.security_guardrail(tech['emp_id'], req_safety):
                    if tech['emp_id'] not in self.state['operator_id'].values:
                        decisions.append({
                            'Action': 'DEPLOY_FLOAT',
                            'Operator': tech['emp_id'],
                            'To_Operation': op_name,
                            'Expected_Boost': f"+{tech[op_name] - 70}%"
                        })
                        break
        return decisions
```

```python
#  Production-Grade Live Dashboard
orchestrator = DenimFlowOrchestrator(line_df, skill_df)

try:
    for i in range(5):  # Simulating 5 cycles of a shift
        clear_output(wait=True)
        print(f"--- DENIM-FLOW AGENT: LIVE FACTORY MONITOR (Cycle {i+1}) ---")
        
        # 1. Run Agents
        bottlenecks = orchestrator.sam_monitor_agent()
        recommendations = orchestrator.workforce_optimizer_agent(bottlenecks)
        
        # 2. Display Status
        print("\n[LIVE EFFICIENCY STATUS]")
        display(orchestrator.state[['operation', 'efficiency', 'operator_id']])
        
        if not bottlenecks.empty:
            print("\n⚠️ BOTTLENECK DETECTED!")
            display(bottlenecks[['operation', 'efficiency']])
            
            print("\n✅ AGENT RECOMMENDATION:")
            for rec in recommendations:
                print(f"👉 {rec['Action']}: Move {rec['Operator']} to {rec['To_Operation']} (Safety Verified)")
        
        # Hardware Optimization: 2-second sleep to keep CPU thermal levels low
        time.sleep(2) 

except KeyboardInterrupt:
    print("\nSimulation halted by user.")
```

```python
#  Predictive Drift Node
def predictive_drift_agent(self, historical_data_list):
    """
    Node 4: Uses Linear Regression to foresee bottlenecks.
    historical_data_list: A list of DataFrames from previous cycles.
    """
    predictions = []
    
    # We need at least 3 cycles to establish a reliable trend
    if len(historical_data_list) < 3:
        return predictions

    # Combine history to track SAM changes over time
    for op in self.state['operation']:
        # Extract the live_sam for this specific operation across all cycles
        y = [df.loc[df['operation'] == op, 'live_sam'].values[0] for df in historical_data_list]
        x = np.array(range(len(y))) # Time steps (0, 1, 2...)
        
        # Polyfit(deg=1) calculates the slope (m) and intercept (b) for y = mx + b
        slope, intercept = np.polyfit(x, y, 1)
        
        # Industrial Logic: If slope > 0, efficiency is decaying (SAM is rising)
        if slope > 0:
            target_sam = self.state.loc[self.state['operation'] == op, 'target_sam'].values[0]
            # Calculate predicted efficiency 3 cycles into the future
            future_sam = slope * (len(y) + 3) + intercept
            future_eff = (target_sam / future_sam) * 100
            
            if future_eff < 85:
                predictions.append({
                    'op': op,
                    'trend': "DEGRADATION",
                    'est_efficiency': round(future_eff, 2)
                })
                
    return predictions

# Attach to Class
DenimFlowOrchestrator.predictive_node = predictive_drift_agent
```

```python
#Sustainable Data Update
def update_data_with_sustainability(line_df):
    """Adds real-time fabric waste tracking to the production line."""
    # Simulating waste in grams per garment (target is < 15g)
    line_df['waste_grams'] = [12.5, 14.2, 22.8, 11.9, 13.5] 
    line_df['waste_threshold'] = 15.0
    return line_df

line_df = update_data_with_sustainability(line_df)
print("Sustainability Metrics Integrated: Carbon & Waste Tracking Active.")
```

```python
#  Sustainability Agent Node
def sustainability_monitor_agent(self):
    """
    Node 5: Detects 'Pseudo-Efficiency' (High speed but high waste).
    Calculates Carbon impact based on fabric loss.
    """
    waste_alerts = []
    
    # Calculate Waste Variance
    self.state['waste_variance'] = self.state['waste_grams'] - self.state['waste_threshold']
    
    # Identify 'Dirty Speeders' (Efficiency > 90% but Waste > Threshold)
    dirty_speeders = self.state[(self.state['efficiency'] > 90) & (self.state['waste_variance'] > 0)]
    
    for _, row in dirty_speeders.iterrows():
        # Calculation: 1kg of wasted Denim ≈ 20kg of CO2 (Approx industrial avg)
        co2_impact = (row['waste_variance'] * 100) * 0.02 
        
        waste_alerts.append({
            'op': row['operation'],
            'emp': row['operator_id'],
            'excess_waste': round(row['waste_variance'], 2),
            'co2_penalty': round(co2_impact, 2),
            'insight': "Fast but exceeding waste limits. Requires machine re-calibration."
        })
        
    return waste_alerts

# Attach to Class
DenimFlowOrchestrator.sustainability_node = sustainability_monitor_agent
```

```python
#  The "What-If" Simulation Engine
def run_what_if_scenario(self, op_to_change, new_operator_id):
    """
    Node 6: Predictive Simulation. 
    Tests the impact of swapping an operator before execution.
    """
    # 1. Create a "Sandbox" (Memory Management)
    sandbox_state = self.state.copy()
    
    # 2. Look up the new operator's skill for that specific operation
    new_skill_row = self.skills.loc[self.skills['emp_id'] == new_operator_id]
    
    if new_skill_row.empty:
        return "Error: Operator ID not found in Skill Matrix."

    # 3. Simulate the change: Update the operator and their live_sam
    # We assume live_sam will improve/decline based on their proficiency score
    new_proficiency = new_skill_row[op_to_change].values[0]
    target_sam = sandbox_state.loc[sandbox_state['operation'] == op_to_change, 'target_sam'].values[0]
    
    # Logic: Predicted SAM = Target / (Proficiency/100)
    predicted_sam = target_sam / (new_proficiency / 100)
    
    # Apply changes to the Sandbox only
    sandbox_state.loc[sandbox_state['operation'] == op_to_change, 'operator_id'] = new_operator_id
    sandbox_state.loc[sandbox_state['operation'] == op_to_change, 'live_sam'] = predicted_sam
    sandbox_state['efficiency'] = (sandbox_state['target_sam'] / sandbox_state['live_sam']) * 100
    
    # 4. Calculate ROI (Return on Investment)
    old_eff = self.state.loc[self.state['operation'] == op_to_change, 'efficiency'].values[0]
    new_eff = sandbox_state.loc[sandbox_state['operation'] == op_to_change, 'efficiency'].values[0]
    
    return {
        'operation': op_to_change,
        'improvement': round(new_eff - old_eff, 2),
        'final_efficiency': round(new_eff, 2),
        'status': "✅ FEASIBLE" if new_eff >= 85 else "❌ RISKY"
    }

# Attach to Class
DenimFlowOrchestrator.simulate_scenario = run_what_if_scenario
```

```python
# Demo: What if we move the high-skilled 'EMP_FLOAT_01' to the struggling 'Inseam' line?
result = orchestrator.simulate_scenario('Inseam', 'EMP_FLOAT_01')

print(f"--- SIMULATION RESULT for {result['operation']} ---")
print(f"Predicted Efficiency Boost: {result['improvement']}%")
print(f"Final Outcome: {result['final_efficiency']}% ({result['status']})")
```
